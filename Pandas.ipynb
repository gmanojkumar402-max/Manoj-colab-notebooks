Inplace parameter
Inplace is a parameter for DataFrame methods like rename(p, inplace), drop(p, inplace), sort_values(p, inplace) and set_index(p, inplace)

The above methods will create a new dataframe whenver they are used instead of making changes on existing DataFrame.

inplace parameter will avoid creating new DataFrame

When we provide inplace parameter as True changes will apply on existing DataFrame

```python
import pandas as pd
a = {
'order_id': [1,2,3,4],
'customer_name': ['M', 'P', 'D', 'P'],
'product': ['iphone','sam','oppo','micromax']
}
col = {'order_id': 'ord_id', 'customer_name':'cus_name', 'product':'product'}
df = pd.DataFrame(a)
df.rename(columns = col, inplace = True)
df
```

Key rule (VERY important)

When using & or | --> or in Pandas:

✅ ALWAYS wrap conditions in parentheses

DataFrame Merging or joining

This is same as join operations in database and here we are performing join in between the DataFrames

merge(p1, p2, p3, p4) is a predefined function in pandas.
o pd.merge(df1, df2, on = "column", how = "type of join")

'how'  is a key word argument
The how argument helps to specify the type of join.

Inner Merge / Inner join
We can perform inner join on DataFrames.

This inner join is a kind of intersection means, it keeps the common data.


```python
# inner join
import pandas as pd

d1 = {
    "Id": [1, 2, 3, 4, 5, 6],
    "Name": ["Pradhan", "Venu", "Madhurima", "Nireekshan", "Shafi", "Veeru"],
    "Subject": ["English", "Java", "Html", "Python", "C", "dot net"]
}

d2 = {
    "Id": [11, 12, 13, 14, 15, 16],
    "Name": ["Srinu", "Sumanth", "Neelima", "Daniel", "Arjun", "Veeru"],
    "Subject": ["Java", "Html", "Cpp", "Python", "C", "dot net"]
}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
inner_join = pd.merge(df1,df2, on = 'Subject', how = 'inner')
inner_join
```

In inner join the common data in both the dataframes we will get it as result if no common it will be left out automatically

Left Merge / Left outer join:

Keep every row in the left dataframe.

Where there are missing values of the "on" variable in the right dataframe filled with NaN values in the result.

We will get left side data along with the common data from right side and wherever the record is not matching on the left side data with right side data there we will get NaN

```python
import pandas as pd

d1 = {
    "Id": [1, 2, 3, 4, 5, 6],
    "Name": ["Pradhan", "Venu", "Madhurima", "Nireekshan", "Shafi", "Veeru"],
    "Subject": ["English", "Java", "Html", "Python", "C", "dot net"]
}

d2 = {
    "Id": [11, 12, 13, 14, 15, 16],
    "Name": ["Srinu", "Sumanth", "Neelima", "Daniel", "Arjun", "Veeru"],
    "Subject": ["Java", "Html", "Cpp", "Python", "C", "dot net"]
}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
left_join = pd.merge(df1,df2, on = 'Subject',how = 'left')
left_join
```

```python
# right join is opposite of left join
right_join = pd.merge(df1, df2, on = 'Subject', how = 'right')
right_join
```

Outer Merge / Full outer join:

A full outer join returns all the rows from the left and right DataFrames.

Where there is no common data there it will be filled with NaN values.


```python
outer_join = pd.merge(df1, df2, on = 'Subject', how = 'outer')
outer_join
```

Many-to-one joins occur when the key column has duplicate values in one DataFrame and unique values in the other.

```python
# Many to one
import pandas as pd

b1 = {
    "Employee": ["Nireekshan", "Veeru", "Lavanya", "Pradhan"],
    "Group": ["Development", "Testing", "Testing", "HR"]
}

b2 = {
    "Employee": ["Lavanya", "Nireekshan", "Veeru", "Pradhan"],
    "Hire_date": [2010, 2012, 2014, 2016]
}

b3 = {'Group': ['Testing', 'Development', 'HR'],
      'supervisor':['Shafi', 'Daniel', 'Neelima']
}
df1 = pd.DataFrame(b1)
df2 = pd.DataFrame(b2)
df3 = pd.DataFrame(b3)
# first let's do one_to_one
one_one = pd.merge(df1,df2)
print(one_one)
print()
print()
print(df3)
```

```python
# Now let's perform many_to one
many_to_one = pd.merge(one_one, df3)
many_to_one
```

Many-to-many joins occur when the key column has duplicate values in both DataFrames, resulting in multiple combinations of matching rows.

```python
# Many_to_Many
import pandas as pd

d1 = {
    "Employee": ["Nireekshan", "Veeru", "Lavanya", "Pradhan"],
    "Group": ["Development", "Testing", "Testing", "HR"]
}

d2 = {
    "Group": ["Testing", "Testing", "Development", "Development", "HR", "HR"],
    "Skills": ["Manual", "Automation", "Coding", "Logical", "Spreadsheets", "Organization"]
}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
print(df1)
print()
print(df2)
many_to_many = pd.merge(df1,df2)
many_to_many
```

Concatenate Multiple csv files

To load all files from a folder we have os module whcih contains a function listdir --> which means from the directory list all the file names.

next we have to create a string wrt foldername

ex: './folder name' --> . indicates that the folder is existing in the current directory

Next assign a variable to the string --> a = './folder name'

final syntax: all_files = listdir(a)

Next step after accessing all the files we have to filter the files based on the categories.

In sting we have a method called endswith and it returns only boolian data type as result


```python
a = ["Naresh","Suresh","Daniel","Phani","Rajesh","Ramesh"]
# results = [p1,p2,p3]
# results = [variable, for loop, condition]
result = [names for names in a if names.endswith('sh')]
result
```

Now we can even use filter concept for the above result

```python
a = ["Naresh","Suresh","Daniel","Phani","Rajesh","Ramesh"]
# f = filter(p1,p2)
# f = filter(lambda function, always existing list data structure)
f = filter(lambda names: names.endswith('sh'), a)
f
# Output is filter object
```

```python
# Now let's convert into list data structure
b = list(f)
b
```

listdir is a predefined function in os module

This function returns all the file names in the lsit

filter(p1, p2) is a predefined function in python and using filter function we can apply Boolean logic and get results.

Concatenating all csv file

Once after loading all csv files using pandas we can concatenate all csv files into one csv file

all_files = os.path.join(always the folder name that we have created , provide the extension of the file that you are intending to filter)

```python
import os
import glob
a = '/content/sample_data/daniel'
all_files = os.path.join(a, '*.csv')
# we have glob function in glob module.
# It takes a single parameter and gets all files with fully qualified path
# fully qualified path  file names along with their location
alll_files = glob.glob(all_files)
alll_files
```

```python
# Now let's print all the files one by one
for files in alll_files:
  print(files)
```

Now let's create a dataframe using the files

We have to make sure that df=pd.read_csv(files) is within the loop orelse only the last file in the loop will get executed leaving other files

All 12 files will be printed

```python
import pandas as pd
for files in alll_files:
  print(files)
  df=pd.read_csv(files)
  print(df)
```

In Real-time for loops are not being used

We will use list comprehension syntax in real time.

all_dfs = [p1, p2]
all_dfs = [variable, for loop]

all 12 files will be printed

```python
all_dfs = [pd.read_csv(files)  for files in alll_files]
print(all_dfs)
```

```python
# Now we shall concatenate all the files.
all_year_files = pd.concat(all_dfs, ignore_index = True)
all_year_files
```

```python
# To get this concatenated DataFrame as a physical file.
all_year_files.to_csv('year.csv')
```
