Numpy is used for scientific computing in python

Numpy provides multidimensional array object

An array is an object that stores a group of values in an ordered manner. These values are typically of the same data type and can be accessed using an index.

Key Characteristics of Arrays

•	Ordered collection → elements are stored in a sequence

•	Indexed → each element has a position (starting from 0 in Python)

•	Same data type → usually stores elements of the same type

•	Efficient storage → especially for numerical data


The array object in numpy is called as ndarray.

ndarray--> N-dimensional (can have 1D, 2D, 3D, or more dimensions) array

```python
# Explanation
empty_list = []
for rows in sales: # We are iterating through each row of the array
  new_rows = []       # creating an empty list for every new row
  for val in rows:  # we are iterating through each value in that row
    new_rows.append(val + 2) # we add val + 2 and store it in new rows
  empty_list.append(new_rows)  # we are adding new rows to empty list
print(empty_list)  # finally printing the empty list
```

Broadcasting

Broadcasting in NumPy allows arithmetic operations between arrays of different shapes by automatically expanding the smaller array to match the larger one, without explicitly copying data.

```python
# Broadcasting
import numpy as np

sales = np.array([
    [100, 200, 150],
    [120, 220, 130],
    [90,  210, 160]
])

sales + 10
```

```python
# Add different bonus per product
bonus = np.array([5,10,15])
print(sales + bonus)
```

```python
# Add bonus per day (row-wise)
bonus = np.array([[5],
                  [10],
                  [15]])
print(sales + bonus)
```

```python
# Broadcasting using numpy interference in python code
import numpy as np

sales = np.array([
    [100, 200, 150],
    [120, 220, 130],
    [90,  210, 160]
])

bonus = [10, 20, 30]
empty_list = []
for rows in sales:
  new_rows = []
  for val in rows:
    new_rows.append(val + bonus)
  empty_list.append(new_rows)
print(empty_list)
```

```python
bonus = [10, 20, 30]

empty_list = []

for row in sales:
    new_row = []
    for i in range(len(row)):
      new_row.append(row[i] + bonus[i])
    empty_list.append(new_row)
print(empty_list)
```

Boolean index

Boolean indexing in NumPy allows filtering or modifying array elements based on conditions. It works by creating a Boolean mask (True/False) and selecting elements where the condition is True.

```python
# Boolean index
# get values > 150
import numpy as np
sales = np.array([10, 20, 30, 40, 50])
print(sales)
```

```python
# creating a condition
sales > 20
```

```python
# To select values where condition is True
sales[sales > 20]
```

```python
# creating a condition
sales < 40
```

```python
# To select values where condition is True
sales[sales < 40]
```

```python
# 2D Example
sales = np.array([
    [100, 200, 150],
    [120, 80, 130],
    [90, 210, 160]
])
sales > 150
sales[sales > 150]
```

```python
# Replace low values
sales[sales < 100] = 0
sales
```

Axis Operations

Axis in NumPy defines the direction along which operations are performed. Axis 0 refers to column-wise operations (down rows), while axis 1 refers to row-wise operations (across columns).

```python
import pandas as pd

sales = np.array([[100, 200, 150],
                  [120, 220, 130],
                  [90, 210, 160]])
sales
```

Axis tells NumPy the direction along which an operation should be performed.

In NumPy:

axis = 0 → column-wise (downwards)

axis = 1 → row-wise (across)

```python
sales.sum(axis = 0)
```

```python
sales.sum(axis = 1)
```

```python
# product wise average
sales.mean(axis = 0)
```

```python
# day wise average
sales.mean(axis = 1)
```

```python
# max value per column
sales.max(axis = 0)
```

```python
# max value per row
sales.max(axis = 1)
```

Normal Distribution

A normal distribution is a type of data distribution where values are spread symmetrically around the mean (average), forming a bell-shaped curve.

```python
# Normal Distribution
data = np.random.randn(3,3)
print(data)
```

Random module is used to Generate dummy data, Train/test split, Initialize ML models and Simulations

The NumPy random module is used to generate random numbers and datasets. It is widely used in data science for simulations, sampling, and machine learning tasks. Functions like rand, randint, choice, and randn help create different types of random data.

```python
data = np.random.randn(3)
print(data)
print()
print(data.min(), data.max())
```

In a normal distribution, values are symmetrically distributed around the mean, meaning the left and right sides have the same shape and probability distribution, but not necessarily the exact same number of values.
