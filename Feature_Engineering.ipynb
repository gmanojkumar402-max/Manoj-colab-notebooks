Featuring engineering --> Data Preperation

In feature engineering, we identify which features are important for predicting the target variable and which are not. For example, in predicting rice production, the area of land is an important feature, while the farmer's name is not relevant. Feature engineering also involves handling missing values, treating outliers, and transforming data to make it suitable for machine learning models.

Feature engineering is the process of selecting, transforming, and creating relevant features from raw data to improve model performance. It includes identifying important features, removing irrelevant ones, handling missing values and outliers, and transforming data into a suitable format.

In feature engineering, we often use libraries where we import classes, create objects, and call methods (especially in scikit-learn), but many operations can also be performed using functions without creating objects.

```python
# LabelEncoder
from sklearn.preprocessing import LabelEncoder
import pandas as pd
data = {'Company': ['Google','Twitter','Google','Linkedin'], 'Role': ['Data scientist','Sales manager','HR','HR']}
df = pd.DataFrame(data)
le = LabelEncoder()
df['Company_n'] = le.fit_transform(df['Company'])
df['Role_n'] = le.fit_transform(df['Role'])
print(df)
```

Impute Missing values

There are two types of missing values

Numeric missing values

Categorical missing values

We can impute missing numeric values with different strategies like mean, median, most_frequent, constant

Imputing missing values refers to the process of replacing missing or null values in a dataset with substitute values.

```python
# Impute Missing values
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
students = [[85, 'M', 'verygood'],
[95, 'F', 'excellent'],
[60, None, 'good'],
[np.nan, 'M', 'average'],
[70, 'M', 'good'],
[np.nan, None, 'verygood'],
[60, 'F', 'verygood'],
[98, 'M', 'excellent']]
cols = ['Marks', 'Gender', 'Result']
df = pd.DataFrame(students, columns = cols)
print(df)
```

np.nan is commonly used to represent missing numerical values, while None is a Python object used to represent missing values in object or string data. In Pandas, both can be used, but np.nan is preferred for numerical operations.

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

The missing_values parameter specifies what constitutes a missing value in our dataset.

The strategy parameter defines how the missing values should be replaced

Note: imputer object works with single column array

```python
# creating a single column array
df['Marks'] # This return a Series
```

```python
df['Marks'].values # This returns a single row array
```

```python
# We don't need single row array. We need a single column array
# reshape method
df['Marks'].values.reshape(8,1)
```

```python
# Now assigning a variable to a single column array
a = df['Marks'].values.reshape(8,1)
# we can even use the below syntax
a = df['Marks'].values.reshape(-1,1)
# -1 -1 tells NumPy to automatically calculate that dimension.
```

```python
# changing the existing column data
df['Marks'] = imputer.fit_transform(a)
print(df)
```

If there is no most frequent value then according to ascending order whatever is the 1st value that value will be considered.

```python
# Imputing missing numeric values with most_frequent strategy
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
students = [
[85, 'M', 'verygood'],
[95, 'F', 'excellent'],
[60, None, 'good'],
[np.nan, 'M', 'average'],
[70, 'M', 'good'],
[np.nan, None, 'verygood'],
[60, 'F', 'verygood'],
[98, 'M', 'excellent']
]
cols = ['Marks','G','result']
df = pd.DataFrame(students, columns = cols)
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
a = df['Marks'].values.reshape(-1,1)
df['Marks'] = imputer.fit_transform(a)
print(df)
```

```python
# Imputing missing numeric values with constant strategy
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
students = [
[85, 'M', 'verygood'],
[95, 'F', 'excellent'],
[60, None, 'good'],
[np.nan, 'M', 'average'],
[70, 'M', 'good'],
[np.nan, None, 'verygood'],
[60, 'F', 'verygood'],
[98, 'M', 'excellent']]
cols = ['Marks','G', 'R']
df = pd.DataFrame(students, columns = cols)
imputer = SimpleImputer(missing_values = np.nan, strategy = 'constant', fill_value = 80)
a = df['Marks'].values.reshape(-1,1)
df['Marks'] = imputer.fit_transform(a)
df
```

Handling categorical values

Categorical data are variables that contain label values rather than numeric values

Nominal Variable
The variables which are having no-order those are called as Nominal Variable.

Ex: colors like blue, green and red

Ordinal Variable
The variables which are having an order those are called as ordinal Variable.

Ex: Score variables values : low, medium, high


There are 3 ways to convert categorical variables to numerical values.

Ordinal encoding

One hot encoding

Dummy variable encoding


Ordinal encoding: In ordinal encoding every nominal value is assigned an integer value for

Ex: blue : 0, green : 1, red : 2

By assigning an integer in alphabetical order to the nominal data we are trying to bring an order or trying to convert to ordinal data

Ordinal encoding is used when:Data has meaningful order like School < College < PhD

if data has no order it should not be used like Red=1, Blue=2, Green=3


Model may assume: Green (3) > Blue (2) > Red (1)

But in reality colors have no order

```python
# Ordinal encoding
# asarray function
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
data = np.asarray([['blue'], ['green'], ['red']])
print(data)
```

```python
encoder = OrdinalEncoder()
result = encoder.fit_transform(data)
print(result)
```

One-Hot Encoding converts nominal categorical data into numerical features without introducing any ordinal relationship between the categories.

OneHotEncoder class encodes or converts categorical data into vectors.

sparse = mostly zeros

sparse=True → saves space by not storing all the zeros → output type is sparse matrix → not very redable

sparse=False → gives full array with 0s and 1s (good for learning/printing) → output type is numpy array → easy to read and understand


```python
# One Hot Encoding
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
a = [['apple'], ['peer'], ['apple'], ['peer'], ['apple']]
result = np.asarray(a)
result
```

```python
OHE = OneHotEncoder(sparse_output = False)
OHE.fit_transform(result)
```

Dummy variable encoding: Apply one hot encoding and drop the first column

array() function always makes a new copy of the data

asarray only makes a copy if needed otherwise, reuses the original if it is already an array. So, asarray is more memory efficient

```python
# Dummy variable encoding
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
data = [['blue'], ['green'], ['red']]
result = np.asarray(data)
print(result)
print()
OHE = OneHotEncoder(sparse_output = False)
result = OHE.fit_transform(result)
print(result)
```

When we use all binary columns in one-hot encoding, one column becomes redundant or of no use because it can be derived from the others.

Ex:

o blue : 1 0 0

o green : 0 1 0

o red : 0 0 1

Dummy variable encoding example

Ex:

blue : 0 0

green : 1 0

red : 0 1


If we drop first column from the result of one hot encoding, then we will get dummy variable encoding.

```python
# Dummy variable encoding
OHE = OneHotEncoder(drop = 'first', sparse_output = False)
result = OHE.fit_transform(result)
result
```

Imputing Missing class values


If qunatitative replace with an average value and Median → better for skewed data and If data has outliers → use median

If qualitative (categorical data) replace it with most common value or Mode (most frequent value)

Advanced methods like KNN imputation can also be used to predict missing values based on other features.

Imputation is the process of replacing missing values with meaningful substitutes instead of removing the data.

Categorical feature may have missing values which we can impute with most frequent strategy

```python
# Imputing Missing Class Values
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
students = [
[85, 'M', 'verygood'],
[95, 'F', 'excellent'],
[75, np.nan, 'good'],
[np.nan, 'M', 'average'],
[70, 'M', 'good'],
[np.nan, np.nan, 'verygood'],
[92, 'F', 'verygood'],
[98, 'M', 'excellent']]
cols = ['marks', 'gender', 'result']
df = pd.DataFrame(students, columns = cols)
print(df)
print()
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
a = df['gender'].values.reshape(-1,1)
b = df['marks'].values.reshape(-1,1)
df['marks'] = imputer.fit_transform(b)
df['gender'] = imputer.fit_transform(a).flatten()
df
```

But df.gender expects a 1D Series. For this we just need to flatten the 2D output using .ravel() or flatten()

.fit_transform(result) → gives shape (n, 1) → 2D

.ravel() or .flatten() → converts it back to 1D shape (n,), which can be safely assigned to df['gender'].

SimpleImputer returns a 2D array. When we manually reshape a column using reshape(-1,1), the output remains 2D, so we use flatten() to convert it back to 1D before assigning it to a DataFrame column. However, when using df[['col']], Pandas handles the conversion automatically, so flattening is not required.
