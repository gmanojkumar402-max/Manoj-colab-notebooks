```python
# Multiple Linear Regression using sklearn

from sklearn.datasets import load_diabetes
import numpy as np
X,y = load_diabetes(return_X_y=True)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 2)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)

# Multiple Linear Regression using our own class


class MLR:
  def __init__(self):
    self.coef_ = None     # initializing coefficient
    self.intercept_ = None # initializing intercept
  def fit(self,X_train,y_train):
    X_train = np.insert(X_train,0,1,axis = 1) # (array under which changes are required, which index changes should reflect,what change we need the value we will put, and we will give the axis)

    # calculate the coeffs
    beta = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
    self.intercept_ = beta[0]
    self.coef_ = beta[1:]
  def predict(self,x_test):
    y_pred = np.dot(X_test,self.coef_) + self.intercept_
    return y_pred
```

```python
mlr = MLR()
```

```python
mlr.fit(X_train,y_train)
```
