# map is a predefined function it takes iterable and apply logic on every item and return iterable

```python
# map function
without_gst_cost = [100,200,300,400]
with_gst_cost = map(lambda x:x+10, without_gst_cost)
x = list(with_gst_cost)
print("without GST item cost:", without_gst_cost)
print("with GST cost",x)
```

```python
cost_of_two_item =[20,30]
cost_of_the_items_after_adding_profit =map(lambda x: x +10,cost_of_two_item)
y = list(cost_of_the_items_after_adding_profit)
print("cost_of_two_item:", cost_of_two_item)
print("cost_of_the_items_after_adding_profit:", cost_of_the_items_after_adding_profit)
```

```python
cost_of_two_item =[20,30]
cost_of_the_items_after_adding_profit =map(lambda x: x +10,cost_of_two_item)
y = list(cost_of_the_items_after_adding_profit)
print("cost_of_two_item:", cost_of_two_item)
print("cost_of_the_items_after_adding_profit:", y)
```

```python
# filter(p1,p2)function. it has same properties as map
items_cost = [1,2,3,4,5,6]
gt_num = filter(lambda x : x > 4, items_cost)
y = list(gt_num)
print("items_cost:", items_cost)
print("eligible for discount:",y)
```

reduce(p1,p2) function
#This function apply a function of two arguments cumulatively to the items of sequence, from left to right.
# Finally this function reduce the sequence to a single value.

```python
# reduce(p1, p2) function
from functools import reduce
items_cost = [1,2,3,4,5]
total_cost = reduce(lambda x,y: x+y, items_cost)
print("items_cost:", items_cost)
print("total_cost:", total_cost)
```

# Functions with NumPy / Pandas style operations

```python
# numeric scale
import pandas as pd
import numpy as np

data = {
    "name": ["A", "B", "C", "D"],
    "age": [22, 17, 35, 29],
    "salary": [30000, 15000, 50000, 40000]
}

df = pd.DataFrame(data)
df

def min_max_scale(series):
  """
  Scale values in a series to between 0 and 1.
  """
  return (series - series.min())/(series.max() - series.min())
df["salary_scaled"] = min_max_scale(df["salary"])
df
```

```python
# row-wise logic
def label_age(age):
  if age< 18:
    return "minor"
  elif age < 30:
    return "young"
  else:
    return "adult"
df["age_group"] = df["age"].apply(label_age)
df
```

```python
# Function to filter data safely
def filter_high_salary(df, threshold):
  return df[df["salary"]> threshold]
high_salary_df = filter_high_salary(df, 30000)
high_salary_df
```

```python
# Function with NumPy logic
def log_transformation(series):
  return np.log(series)
df["log_salary"] = log_transformation(df["salary"])
df
```

```python
# Small functional pipeline (real DS flow)
def clean_salary(df):
  return df[df["salary"]>0]
def add_log_salary(df):
  df = df.copy()
  df["log_salary"] = np.log(df["salary"])
  return df
df_pipeline = add_log_salary(clean_salary(df))
df_pipeline
```

Small functional pipelines (VERY IMPORTANT)

```python
# simple pipeline
def clean(data):
  return  [x for x in data if x is not None]
def square(data):
  return [x**2 for x in data]
c = square(clean([1, None, 2, 3]))
print(c)
```

```python
# feature pipeline
def add_tax(price):
  return price * 1.05
def discount(price):
  return price * 0.9
x = discount(add_tax(100))
print(x)
```

```python
# EDA-style pipeline
def remove_negatives(nums):
  return [x for x in nums if x >=0]
def avg(nums):
  return sum(nums)/len(nums)
x = avg(remove_negatives([1,2,3,4,-5]))
print(x)
```
