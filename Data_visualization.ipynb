Pie chart:

A pie chart is a circular graph divided into slices, where each slice represents the proportion of a category to the whole. Larger values occupy larger portions of the chart. The command plt.axis('equal') ensures that the pie chart is drawn as a perfect circle by maintaining equal scaling on both axes.

```python
# Without plt.axis('equal')
import matplotlib.pyplot as plt
students = ['Daniel','Prasad','Srinu']
points=[62,48,36]
plt.pie(points, labels = students)
plt.show()
```

```python
import matplotlib.pyplot as plt
students = ['Daniel','Prasad','Srinu']
points=[62,48,36]
plt.pie(points, labels = students)
plt.axis('equal')
plt.show()
```

What does plt.axis('equal') do?
When you plot something in Matplotlib, it automatically scales the axes to fit the data. However, this scaling might stretch or shrink the plot, making circles look like ovals or squares look like rectangles.
Using plt.axis('equal'), we tell Matplotlib:
"Make sure that 1 unit on the x-axis is the same length as 1 unit on the y-axis."

Final Exam Version
legend() → displays labels to identify different data series.

explode → separates slices in a pie chart.

shadow → adds shadow effect to the pie chart.

autopct → displays percentage values on pie chart slices

The first argument is a list of numerical values. These values represent the size of each category

labels → names of categories

colors → colors for slices

shadow → adds shadow effect

explode → separates slices

% → formatting symbol,
.1f → show 1 decimal place,
%% → print a % sign. So this displays values like 25.3%

autopct = '%.0f%%' --> Output: 25%

autopct = '%.2f%%' --> Output: 25.34%



```python
# Pie chart
import matplotlib.pyplot as plt
students =['daniel','prasad','srinu']
points=[62,48,36]
c=['y','r','b']
plt.pie(points, labels = students, explode = [0.05,0.05,0.05], colors = c, shadow = True, autopct = '%.1f%%')
plt.axis('equal')
plt.legend()
plt.show()
```

Heat Map

Heat map is also called as a grey scale

imshow() is a function used to display 2D data as an image
When used with a DataFrame, it behaves like a heatmap
Each value is represented by a color

cmap is a keyword argument that defines how values map to colors
It uses a color gradient. ex: "YlGnBu"

Low values → Yellow. Medium values → Green. High values → Blue

plt.colorbar() → Adds a color scale bar beside the plot
Shows which color corresponds to which value

Axis Labels using DataFrame:

df.columns → labels for X-axis

df.index → labels for Y-axis

Tick Functions

plt.xticks() → Sets labels for the X-axis

plt.yticks() → Sets labels for the Y-axis

Tick Positions:

range(len(df)) → Used to create positions for ticks

Why?

imshow() uses numeric positions (0,1,2,...). So, we map them to actual labels

range(len(df.columns)) → For columns (X-axis)

range(len(df)) → For rows (Y-axis)

```python
# Heat Map
import pandas as pd
import matplotlib.pyplot as plt
d = {
        "Apple": [10, 20, 30, 40],
        "Orange": [7, 14, 21, 28],
        "Banana": [55, 15, 8, 12],
        "Pear": [15, 14, 1, 8]
}
i = ['Basket1', 'Basket2', 'Basket3', 'Basket4']
df = pd.DataFrame(d, index = i)
df
```

```python
# Length function
len(df)
```

```python
# Range function
range(len(df))
```

```python
# using imshow() function to display 2D data as an image
plt.imshow(df, cmap = 'YlGnBu')
plt.colorbar() # To add a color scale
plt.xticks(range(len(df)), df.columns)
plt.yticks(range(len(df)), df.index)
plt.show()
```

```python
# Plot the survival rate of each class
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = sns.load_dataset('titanic')
sns.barplot(x = 'pclass', y = 'survived', data = df)
```

By default, Seaborn’s barplot() computes the mean of the y values. Since the survived column contains binary values (1 for survived and 0 for not survived), the mean of survived for each category represents the survival rate.

Survival Rate = Total Survivors in Class / Total Passengers in Class

```python
# We can even write the above code manually
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = sns.load_dataset('titanic')
# calculationg survival rate manually
survival_rate = df.groupby('pclass')['survived'].mean().reset_index()
sns.barplot(x = 'pclass', y = 'survived', data = survival_rate)
plt.title('survival rate by passenger')
```

A distribution plot is used to visualize how numerical data is distributed. It is commonly represented using a histogram, where the x-axis shows data values and the y-axis shows their frequency.

```python
# Dist Plot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = sns.load_dataset('titanic')
sns.distplot(df['fare'])
```

```python
# Histogram
sns.histplot(df['fare'])
# we can see most of the tickets were sold between 0 = 50 dollars
```

A box plot is used to visualize the distribution of numerical data using quartiles (Q1, Q2, Q3). The box represents the range between Q1 and Q3 (interquartile range).The center of the box represents the median value, and it helps in understanding the spread and outliers in the data.

```python
# Box Plot
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = sns.load_dataset('titanic')
sns.boxplot(y = 'age',data = df) # Boxplot takes only series and not df
```

25% of passengers have age ≤ 20

50% of passengers have age ≤ 28

75% of passengers have age ≤ 37

Middle 50% of passengers are between 20 and 37

outliers above 65

Only one numerical column (age) is provided

No categorical grouping (x) is given

Seaborn creates a single box plot

It represents the distribution of the entire age column

The x-axis does not show categories (just one box)

If no categorical variable is provided (i.e., x is not specified), the box plot displays the distribution of a single numerical variable. It shows a single box representing the overall distribution of that variable.

 No x → one box

With x → multiple boxes (grouped)

```python
# Box Plot
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = sns.load_dataset('titanic')
sns.boxplot(x = 'sex',y = 'age',data = df) # Boxplot takes only series and not df
```
