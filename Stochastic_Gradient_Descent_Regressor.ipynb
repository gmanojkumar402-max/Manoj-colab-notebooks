Lets create our own class Stochastic Gradient Descent

```python
class SGDRegressor:

    def __init__(self, learning_rate=0.001, epochs=1000):
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X_train, y_train):
        n_samples, n_features = X_train.shape

        # initialize weights
        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0

        for _ in range(self.epochs):
          for j in range(X_train.shape[0]):
            # we are generating a random number
            random_index = np.random.randint(0,X_train.shape[0])
            # for perticular index lets calculate the prediction
            # in BGD result of y_hat would be a matrix and in our case it is a scalar
            # prediction of a particular index
            y_hat = np.dot(X_train[random_index], self.coef_) + self.intercept_

            # error
            error = y_hat - y_train[[random_index]]

            # gradients (derivatives of intercept and coefficient)
            dW = (2) * X_train[random_index]* error
            dB = (2) * (error)

            # update
            self.coef_ -= self.lr * dW
            self.intercept_ -= self.lr * dB

    def predict(self, X_test):
        return np.dot(X_test, self.coef_) + self.intercept_
from sklearn.preprocessing import StandardScaler

# scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

sgd = SGDRegressor(epochs=50, learning_rate=0.01)
sgd.fit(X_train_scaled, y_train)

y_pred = sgd.predict(X_test_scaled)

print("SGD R2:", r2_score(y_test, y_pred))

# Compare with sklearn
LR = LinearRegression()
LR.fit(X_train, y_train)
print("Sklearn R2:", LR.score(X_test, y_test))

```
