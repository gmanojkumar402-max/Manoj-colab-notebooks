```python
# Inside instance method we can access instance variables by using self.
class Manoj:
  def __init__(self, name):
    self.name = name # instance variable
  def display(self):
    print(self.name)
    print("good boy")
m = Manoj("manoj")
m.display()
```

```python
class Manoj:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def display(self):
    print("name:", self.name)
    print("age:", self.age)
m = Manoj("manoj", 29)
m.display()
```

```python
# real world production quality code
class Manoj:
  def __init__(self, name, age,number):
    self.name = name
    self.age = age
    self.number = number
  def display(self):
    print("name:",self.name)
    print("age:", self.age)
    print("number:", self.number)
m = Manoj("manoj", 29, 143)
m.display()
```

Instance method


Instance methods are methods which act upon the instance variables of
the class.

Instance methods are bound with instances or objects, that’s why called
as instance methods.

The first parameter for instance methods is self variable.

```python
class Manoj:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def display(self):
    print("name:", self.name)
    print("name:", self.age)
m = Manoj("manoj", 29)
m.display()
```
