Data structures are also called as sequence which contain group of elements

List comprehension:

List comprehension takes input as iterable, we can apply conditional
logic on every item and returns new list.


list = [expression for item1 in iterable1 if statement]

Here Iterable represents a list, set, tuple, dictionary or range object.

The result of list comprehension is new list based on the applying
conditions.

```python
# list comprehension example
a = [1,2,3,4,5]
result = [values + 2 for values in a]
print(result)
```

```python
a = [10,20,30,40]
result = [values + 2 for values in a if values >= 20]
print(result)
print(a)
```

```python
a = range(1,11)
results = [values*2 for values in a if values > 1]
print(results)
```

Dictionary data structure

If we want to represent a group of objects or elements as key-value pairs then we
should go for:

o dict or dictionary

We can create dict by using,
Curly braces {} symbols
and dict() predefined function.

Dictionary data structure contains key, value pairs and are separated by colon : symbol and each  key-value pair is called as item and every item is seperated by a comma symbol

In dictionary duplicate keys are not allowed.

In dictionary duplicate values can be allowed.
A dictionary keys and values can store both Homogeneous and Heterogeneous type of elements

In dictionary insertion order is not preserved means not fixed.


Dictionary size will increase dynamically.


Dictionary object having mutable nature.


Dictionary data structure cannot store the elements in index order.So, indexing and slicing concepts are not applicable

```python
# Creating dictionary
a = {10 : "manoj", 20: "vinay", 30: "priya"}
print(a)
```

```python
# Empty dictionary
d = {}
print(d,type(d))
```

```python
# We can add key-value pairs to empty dictionary
d = {}
d[10] = "manoj"
d[20] = "vinay"
d[30] = "priya"
print(d)
```

```python
# Access values by using keys from dictionary
d = {10: "manoj", 20: "vinay", 30: "priya"}
print(d[10])
```

```python
# Accessing key and value from dictionary using for loop
d = {10: "manoj", 20: "vinay", 30: "priya"}
for values in d:
  print(values)
```

```python
# Accessing key and value from dictionary using for loop
d = {10: "manoj", 20: "vinay", 30: "priya"}
for values in d:
  print(values,d[values])
```

While updating the key in dictionary, if key is already existing then old
value will be replaced with new value.

```python
# Update dictionary
d = {10: "manoj", 20: "vinay", 30: "priya"}
d[40] = "shero"
print(d)
```

```python
# Update dictionary
d = {10: "manoj", 20: "vinay", 30: "priya"}
d[10] = "manoj kumar"
print(d)
```

By using del keyword it deletes entry associated with the specified key and if the key is not available, then we will get KeyError.

```python
# Removing or deleting elements from dictionary
d = {10: "manoj", 20: "vinay", 30: "priya"}
del d[10]
print(d)
```

We can delete total dictionary object and once deleted we cannot access the dictionary

```python
d = {10: "manoj", 20: "vinay", 30: "priya"}
print(d)
del d
print(d)
```

```python
# len(p) function
d = {10: "manoj", 20: "vinay", 30: "priya"}
print(len(d))
```

Important methods:
clear() method,

values() method,items() method

A dictionary comprehension represents creating new dictionary from
Iterable object like a list, set, tuple, dictionary and range.

```python
# Dictionary Comprehension:
d = {30: "manoj", 20: "vinay", 10: "priya"}
result = {items for items in d.items()}
print(result)
```

```python
# To return dictionary
d = {30: "manoj", 20: "vinay", 10: "priya"}
result = {k: v for k, v in d.items()}
print(result)
```

List, set, and dictionary are mutable; tuple is immutable.

List, tuple, and dictionary preserve insertion order; set does not.

List, tuple, set, and dictionary can store both homogeneous and heterogeneous data.

Duplicates are allowed in list and tuple; not allowed in set and dictionary keys.

List and tuple support indexing; set does not; dictionary accesses elements using keys.
