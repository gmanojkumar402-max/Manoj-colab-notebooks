```python
# Linear regression using Gradient decent
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
X,y = make_regression(n_samples = 4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state = 13)
plt.scatter(X,y)
```

# Let's apply OLS so that we will have the value of slope and intercept


```python
from sklearn.linear_model import LinearRegression
linearregression = LinearRegression()
linearregression.fit(X,y)
```

```python
linearregression.coef_
```

```python
linearregression.intercept_
```

Now let's write a code where my value of b starts with a random value

```python
plt.scatter(X,y)
plt.plot(X,linearregression.predict(X),color = "red")
```

lets apply Gradient decent assuming slpoe is constant m = 78.35 and starting value of the intercept (b)= 0

```python
y_pred = ((78.35*X) + 0).reshape(4) # our 1st random line away from the best_fit_line
```

lets plot this random line along with the best fit line

```python
plt.scatter(X,y)
plt.plot(X,linearregression.predict(X), color = "green", label = "OLS")
plt.plot(X,y_pred,color = "red", label = "Gradient decent")
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
# let's calculate slope at b = 0
slope =  -2 * np.sum(y - m*X.ravel() - b) # our shape of x (4,1) and y (4,) so we applied X.ravel which converts (4,1) into(4,)
slope
```

```python
# let's assume my learning rate as 0.1
lr = 0.1
step_size = lr * slope
step_size
```

```python
# now lets calculate our new b
b = b - step_size
b
```

As we have our new b let's again calculate y_pred1

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,linearregression.predict(X),color = "red", label ="OLS")
plt.plot(X,y_pred1,color = "green", label = "b = {}".format(b))
plt.plot(X,y_pred,color = "blue",label = "b = 0")
plt.legend()
plt.show()
```

```python
# Interation now at new b point we are again calculating the slope
slope = -2 * np.sum(y - m*X.ravel() -b)
slope
```

```python
step_size = slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,linearregression.predict(X),color = "red", label ="OLS")
plt.plot(X,y_pred2,color = "orange", label = "b = {}".format(b))
plt.plot(X,y_pred1,color = "green", label = "b = {}".format(b))
plt.plot(X,y_pred,color = "blue",label = "b = 0")
plt.legend()
plt.show()
```

Lets take one more iteration

```python
slope = -2 * np.sum(y - m*X.ravel()-b)
slope
```

```python
step_size = slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,linearregression.predict(X),color = "red", label ="OLS")
plt.plot(X,y_pred3,color = "purple", label = "b = {}".format(b))
plt.plot(X,y_pred2,color = "orange", label = "b = {}".format(b))
plt.plot(X,y_pred1,color = "green", label = "b = {}".format(b))
plt.plot(X,y_pred,color = "blue",label = "b = 0")
plt.legend()
plt.show()
```

```python
# instaed of calculation each iteration one by one let's do it in a loop
b = 0
m = 78.35
lr = 0.1
epochs = 10
for i in range(epochs):
  slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b -(lr * slope)
  y_pred = m * X + b
  plt.scatter(X,y)
  plt.plot(X,y_pred)
```

If learning rate is small algorithm may take time but it will rach towards best fit line but if the learning rate is big algorithm may fail to converge
